```python
from collections import deque

def find_buildings(grid, h, w):
    buildings = {}
    visited = set()
    for r_start in range(h):
        for c_start in range(w):
            if (r_start, c_start) in visited or grid[r_start][c_start] == 0:
                continue
            core_color = grid[r_start][c_start]
            q = deque([(r_start, c_start)])
            visited.add((r_start, c_start))
            min_r, max_r, min_c, max_c = r_start, r_start, c_start, c_start
            while q:
                r, c = q.popleft()
                min_r, max_r, min_c, max_c = min(min_r, r), max(max_r, r), min(min_c, c), max(max_c, c)
                for dr, dc in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
                    nr, nc = r + dr, c + dc
                    if 0 <= nr < h and 0 <= nc < w and (nr, nc) not in visited and grid[nr][nc] == core_color:
                        visited.add((nr, nc)); q.append((nr, nc))
            
            is_building = True
            if not (0 < min_r and max_r < h - 1 and 0 < min_c and max_c < w - 1): is_building = False
            if is_building:
                perimeter = [(min_r-1, c) for c in range(min_c - 1, max_c + 2)] + \
                            [(max_r+1, c) for c in range(min_c - 1, max_c + 2)] + \
                            [(r, min_c-1) for r in range(min_r, max_r + 1)] + \
                            [(r, max_c+1) for r in range(min_r, max_r + 1)]
                border_color = grid[perimeter[0][0]][perimeter[0][1]]
                if border_color == core_color or border_color == 0 or any(grid[r][c] != border_color for r, c in perimeter):
                    is_building = False
            if is_building:
                buildings[core_color] = {'core': core_color, 'border': border_color, 'bbox': (min_r-1, max_r+1, min_c-1, max_c+1)}
    return buildings

def solve(grid):
    out = [row[:] for row in grid]
    h, w = len(grid), len(grid[0])
    p_idx = next((r for r in range(h-1, -1, -1) if any(val > 0 for val in grid[r])), -1)
    order = []
    if p_idx != -1:
        for c in grid[p_idx]:
            if c > 0 and c not in order: order.append(c)
    b_map = find_buildings(grid, h, w)
    b_ord = [b_map[c] for c in order if c in b_map]
    if not b_ord: return out
    u_borders = {b['border'] for b in b_ord}
    use_core = len(u_borders) == 1 and list(u_borders)[0] != 1
    for i in range(len(b_ord) - 1):
        b1, b2 = b_ord[i], b_ord[i+1]
        p_color = b2['core'] if use_core else b2['border']
        r1_min, r1_max, c1_min, c1_max = b1['bbox']
        r2_min, r2_max, c2_min, c2_max = b2['bbox']
        c1_mid, r1_mid = (c1_min + c1_max) // 2, (r1_min + r1_max) // 2
        c2_mid, r2_mid = (c2_min + c2_max) // 2, (r2_min + r2_max) // 2
        dr, dc = r2_mid - r1_mid, c2_mid - c1_mid
        p1 = (r1_mid, c1_max + 1 if dc >= 0 else c1_min - 1)
        p2 = (r2_mid, c2_min - 1 if dc >= 0 else c2_max + 1)
        if abs(dr) > abs(dc):
            p1 = (r1_max + 1 if dr >= 0 else r1_min - 1, c1_mid)
            p2 = (r2_min - 1 if dr >= 0 else r2_max + 1, c2_mid)
        p1r, p1c, p2r, p2c = *p1, *p2
        cr, cc = (p1r, p2c) if b1['border'] % 2 != 0 else (p2r, p1c)
        for r in range(min(p1r,cr), max(p1r,cr)+1): out[r][p1c] = p_color
        for c in range(min(p1c,cc), max(p1c,cc)+1): out[p1r][c] = p_color
        for r in range(min(cr, p2r), max(cr,p2r)+1): out[r][p2c] = p_color
        for c in range(min(cc,p2c), max(cc,p2c)+1): out[cr][c] = p_color
    return out
```
